```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
X, Y = make_circles(n_samples=200, noise=0.1, factor=0.3)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(X[Y == 0, 0], X[Y == 0, 1], color='blue')
plt.scatter(X[Y == 1, 0], X[Y == 1, 1], color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Our Data')
plt.show()
```

```python
import pandas as pd
```

```python

df_X = pd.DataFrame(X)
df_X.to_csv('circles_X.csv', index=False, header=False)

df_y = pd.DataFrame(Y)
df_y.to_csv('circles_y.csv', index=False, header=False)
```

```python
X_test, Y_test = make_circles(n_samples=50, noise=0.1, factor=0.3)
df_X = pd.DataFrame(X_test)
df_X.to_csv('circles_X_test.csv', index=False, header=False)

df_y = pd.DataFrame(Y_test)
df_y.to_csv('circles_y_test.csv', index=False, header=False)
```
